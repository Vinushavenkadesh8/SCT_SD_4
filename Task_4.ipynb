```python
import requests
from bs4 import BeautifulSoup
import csv

# 1. Set the URL of the website you want to scrape
url = "http://books.toscrape.com/"

# 2. Fetch the page content
headers = {"User-Agent": "Mozilla/5.0"} # Mimics a real browser
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # 3. Find the product containers (this varies by website)
    products = soup.find_all('article', class_='product_pod')

    scraped_data = []

    for product in products:
        # Extract Name
        name = product.h3.a['title']

        # Extract Price
        price = product.find('p', class_='price_color').text

        # Extract Rating (stored in the class name)
        rating_classes = product.find('p', class_='star-rating')['class']
        rating = rating_classes[1] # e.g., 'Three'

        scraped_data.append([name, price, rating])

    # 4. Save data to a CSV file
    with open('products.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Product Name', 'Price', 'Rating']) # Header
        writer.writerows(scraped_data)

    print("Success! Data saved to products.csv")
else:
    print(f"Failed to retrieve page. Status code: {response.status_code}")
```
